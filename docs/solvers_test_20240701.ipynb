```python
import sys
sys.path.append('../')

from sphere.model.implementations.SIR import SIRModel
from sphere.model.abstract.parameters import SIRParameters
from sphere.output.implementations.SIR import SIROutput
from sphere.model.implementations.solvers.euler import EulerSolver

sir_params = SIRParameters(
    initial_state=(0.99, 0.01, 0.0),
    beta = 0.2,
    gamma = 0.08
)

solver = EulerSolver(delta_t=1.0)
model = SIRModel(sir_params, solver)
output = SIROutput()
output.states = model.run(x0=sir_params.initial_state, t0=0, t_final=100)

output.plot_states()
```
