We added a Transition object that is passed into a Solver upon instantiation. 

The Transition object defines the state transition function for stepping a system 
forward for one time step.

All implementations of a Transition object (SIRTransition, Lorenz63Transition, ...) should 
have a determinstic function. Other models may add a stochastic component as well. 
This would be used in Euler-Maruyama or other SDE solvers. 

```python
from sphere.rume.rume import Rume
from sphere.model.abstract.parameters import SIRParameters

sir_rume = Rume.create_sir_rume(parameters=SIRParameters(beta=0.2, gamma=0.08))
```

```python
sir_rume.run(x0=(0.98, 0.02, 0.0), t0=0, t_final=100)
```

```python
sir_rume.output.plot_states()
```
