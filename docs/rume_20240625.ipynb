We implemented a Runnable Modeling Experiment (RUME) class that wraps around other classes and simplifies the user experience. 

The solver currently defaults to JAXSolver(), but can also be chosen by the user at instantiation. Note that each Rume constructor method could have a different default solver, if we find that one solver works better than others for a specific type of model.

```python
from sphere.parameters.parameters import LorenzParameters
from sphere.rume.rume import Rume


my_rume = Rume.create_lorenz63_rume(
    parameters=LorenzParameters(
        sigma=10.0, rho=28.0, beta=2.5, initial_state=(1.0, 1.0, 1.0)
    )
)

# Alternate Instantiation Method
my_params = LorenzParameters(
    sigma=10.0, rho=28.0, beta=2.5, initial_state=(1.0, 1.0, 1.0)
)
test_rume = Rume.create_lorenz63_rume(parameters=my_params)
```

```python
my_rume.run(time_steps=50)
```

```python
my_rume.output.plot_states()
```

```python
from sphere.parameters.parameters import SIRParameters

sir_rume = Rume.create_sir_rume(
    parameters=SIRParameters(beta=0.25, gamma=0.08, initial_state=(0.98, 0.02, 0.0))
)
```
